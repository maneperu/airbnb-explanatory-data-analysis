#‚≠ê**TOP 5 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE**

```python
# Group by neighborhood and calculate the median price
temp = df.groupby(['neighbourhood'])['price'].median().reset_index()

# Sort neighborhoods based on median price in descending order
temp = temp.sort_values(by='price', ascending=False)

# Select the top 5 expensive neighborhoods
top_5_expensive = temp.head(5)

# Print the top 5 expensive neighborhoods
print(top_5_expensive)
```

```python
# Create a bar plot using Seaborn's barplot
sns.barplot(x="price", y="neighbourhood", data=top_5_expensive, label="Total", color='pink').set(xlabel='Average Price')

# Set the title for the plot
plt.title('TOP 10 EXPENSIVE NEIGHBOURHOOD BASED ON THE AVERAGE PRICE')

# Display the plot
plt.show()

```
